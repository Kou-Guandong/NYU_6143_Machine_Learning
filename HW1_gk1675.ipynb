#### Validation

```python
import numpy as np

n, d, m = 10, 3, 5
x = np.random.randint(1, 100, size=(n, d))
y = np.random.randint(1, 100, size=(m, d))
dist = np.zeros((n, m))
for i in range(n):
    for j in range(m):
        for k in range(d):
            dist[i,j] += (x[i, k] - y[j, k])**2
np.sum((np.expand_dims(x, axis=1) - np.expand_dims(y, axis=0)) ** 2, axis=2) == dist
```
