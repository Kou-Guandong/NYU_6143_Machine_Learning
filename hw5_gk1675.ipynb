## Problem 5

Prepare synthesized data

```python
import numpy as np
```

```python
X = np.random.random((100, 1)) * 5
alpha = np.random.random((10, 1)) * 3
beta = np.random.random((10,1))
y = np.exp(-X.dot(alpha.T)).dot(beta)
```

- Create p = 100 values of  j uniformly in some interval  j  [a,b] where a and b are given.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
```

```python
xtr, xts, ytr, yts = train_test_split(X, y, test_size=0.3)
```

```python
a, b = -10, 10
p = np.linspace(a, b, 100)
```

```python
Ztr = np.exp(-p * xtr)
Zts = np.exp(-p * xts)
```

- Fit the linear model (3) on the training data for some given lam.

```python
lam = 1
model = Lasso(alpha=lam)
model.fit(Ztr, ytr)
beta = model.coef_.T
```

- Measure the test error.

```python
yts_hat = model.predict(Zts)
err_ts = np.sum((yts_hat - yts) ** 2) / len(yts)
```
