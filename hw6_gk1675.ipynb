### 3d)

The new parameter $\alpha$ does not change $\hat{y}$, because the sign of z is not changed. 

$$
z' = W'x + b' = \alpha Wx + \alpha b = \alpha(Wx +  b ) = \alpha z
$$

However, $\alpha$ does the likelyhood

$$
sigmoid(x) = \frac{1}{1+e^{-\alpha z}} 
$$

if z > 0, the larger $\alpha$, the larger the likelyhood;    
if z < 0, the larger $\alpha$, the smaller the likelyhood;
